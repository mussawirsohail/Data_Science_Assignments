```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
gdp = pd.read_csv("GDP.csv")
gdp.head()
```

```python
print("Shape:", gdp.shape)
gdp.info()
gdp.isnull().sum()
gdp.describe()
```

```python
gdp.columns
```

```python
gdp.drop_duplicates(inplace=True)
gdp.fillna(0, inplace=True)
years = [str(y) for y in range(1960, 2023)]
gdp[years] = gdp[years].apply(pd.to_numeric, errors='coerce')
gdp.info()
```

```python
#1 Top 10 countries by GDP in 2022
top10_2022 = gdp[['Country', '2022']].sort_values(by='2022', ascending=False).head(10)

#2 Average GDP per country (already given but recompute to show)
gdp['calculated_avg_gdp'] = gdp[years].mean(axis=1)

#3 GDP growth from 2021‚Äì2022 (recalculate)
gdp['calculated_growth_2021_2022'] = ((gdp['2022'] - gdp['2021']) / gdp['2021']) * 100

#4 Identify countries with negative growth
negative_growth = gdp[gdp['calculated_growth_2021_2022'] < 0][['Country', 'calculated_growth_2021_2022']]

#5 Calculate global total GDP in 2022
total_gdp_2022 = gdp['2022'].sum()
print("Total Global GDP (2022):", round(total_gdp_2022/1e12, 2), "Trillion USD")
```

```python
plt.figure(figsize=(10,6))
sns.barplot(data=top10_2022, y='Country', x='2022', palette='Blues_r')
plt.title("Top 10 Countries by GDP in 2022")
plt.xlabel("GDP (in USD)")
plt.ylabel("Country")
plt.show()
```

```python
top_growth = gdp.sort_values(by='calculated_growth_2021_2022', ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_growth, y='Country', x='calculated_growth_2021_2022', palette='Greens')
plt.title("Top 10 Countries by GDP Growth (2021‚Äì2022)")
plt.xlabel("Growth (%)")
plt.ylabel("Country")
plt.show()
```

```python
countries_to_plot = ['United States', 'China', 'India']

plt.figure(figsize=(12,6))
for c in countries_to_plot:
    country_data = gdp[gdp['Country'] == c][years].T
    plt.plot(country_data.index, country_data.values, label=c)

plt.title("GDP Growth Trend (1960‚Äì2022)")
plt.xlabel("Year")
plt.ylabel("GDP (in USD)")
plt.legend()
plt.xticks(rotation=45)
plt.show()
```

```python
recent_years = ['2018','2019','2020','2021','2022']
plt.figure(figsize=(6,4))
sns.heatmap(gdp[recent_years].corr(), annot=True, cmap='coolwarm')
plt.title("GDP Correlation Between Recent Years")
plt.show()
```
