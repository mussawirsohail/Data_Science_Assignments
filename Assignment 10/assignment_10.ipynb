```python
#61
import matplotlib.pyplot as plt
import numpy as np

# Data
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

# Simple line plot
plt.plot(x, y)
plt.show()
```

```python
#62
age_groups = ['10-20', '21-30', '31-40', '41-50']
students = [15, 40, 25, 10]
plt.bar(age_groups, students, color='skyblue')
plt.title("Number of Students by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Number of Students")
plt.show()
```

```python
#63
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='purple')
plt.title("Random Scatter Plot")
plt.xlabel("X Values")
plt.ylabel("Y Values")
plt.show()
```

```python
#64
data = np.random.randn(1000)

plt.hist(data, bins=30, color='orange', edgecolor='black')
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

```python
#65
categories = ['Category A', 'Category B', 'Category C']
values = [20, 30, 50]

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title("Category Distribution")
plt.show()
```

```python
#66
fig, axes = plt.subplots(1, 2, figsize=(10,4), sharex=True, sharey=True)

axes[0].plot(x, y, color='blue')
axes[0].set_title("Line Plot")

axes[1].scatter(x, y, color='green')
axes[1].set_title("Scatter Plot")

plt.suptitle("Subplots Example: Line vs Scatter")
plt.show()
```

```python
#67
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y, color='red', marker='o', linestyle='--')
plt.title("Customized Line Plot")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
```

```python
#68
categories = ['Category A', 'Category B', 'Category C']
values = [20, 30, 50]

plt.pie(values, labels=categories, autopct='%1.1f%%', startangle=90, colors=sns.color_palette("pastel"))
plt.title("Pie Chart (Seaborn Colors)")
plt.show()
```

```python
#69
data = sns.load_dataset("iris")

sns.scatterplot(data=data, x="sepal_length", y="sepal_width", hue="species", palette="viridis")
plt.title("Seaborn Scatter Plot (Iris Dataset)")
plt.show()
```

```python
#70
sns.lineplot(data=data, x="sepal_length", y="sepal_width", hue="species")
plt.title("Seaborn Line Plot (Iris Dataset)")
plt.show()
```

```python
#71
sns.histplot(data["petal_length"], bins=20, kde=True, color='teal')
plt.title("Seaborn Histogram of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Frequency")
plt.show()
```
